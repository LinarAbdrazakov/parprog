```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# TASK 1
task1 = pd.read_csv("results/results_1.txt", sep='\t', header=None).groupby(0)

mean = np.array(task1.mean())
S = mean[0] / mean

plt.plot(np.arange(1, 9), S)
plt.grid()
plt.title('TASK 1')
plt.ylim(0)
```
